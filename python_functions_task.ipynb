Exercise 4: Lambda Function to Check if value is in a List
Given a list, the task is to write a Python program to check if the value exists in the list or not using the lambda function.


Input  : L = [1, 2, 3, 4, 5]
          element = 4
Output :  Element is Present in the list

Input :  L = [1, 2, 3, 4, 5]
          element = 8
Output :  Element is NOT Present in the list

```python
check_element= lambda L, element: 'Element is present in the list' if list(filter(lambda x: x==element, L))  else "Element NOT present in the list"

L1 = [1,2,3,4,5]
element1 = 4
print(check_element(L1, element1))

L2 = [1,2,3,4,5]
element2 = 8
print(check_element(L2, element2))
```

Exercise 6 :
Write a python function, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. Return the results as a list


```python
def find_even_digit_numbers():
  result = []
  for num in range(1000,3000):
    all_even_digits = all(int(digit)%2==0 for digit in str(num))
    if all_even_digits:
      result.append(num)
  return result

even_digit_numbers = find_even_digit_numbers()
print(even_digit_numbers)
```

Exercise 7 :
Write a python function that accepts a sentence and calculate and return the number of letters and digits.
Suppose the following input is supplied to the program:
hello world! 123
Then, the output should be:
LETTERS 10
DIGITS 3


```python
def count_letters_and_digits(sentence):
  letters_count = sum(c.isalpha() for c in sentence)
  digits_count = sum(c.isdigit() for c in sentence)

  return f"LETTERS {letters_count} \nDIGITS {digits_count}"

input_sentence = "hello world! 123"
result = count_letters_and_digits(input_sentence)
print(result)
```

Exercise 8  MAP:
Write a Python program to convert all the characters into uppercase and lowercase and eliminate duplicate letters from a given sequence. Use the map() function


```python
def change_cases_char(s):
  return str(s).upper(), str(s).lower()

characs = {"e", "u", "B", "G", "a", "E", "T", "S", "e"}
print("Original charecters are:\n", characs)

result = map(change_cases_char, characs)

print("\nAfter converting the above characters in upper and lower cases \n and eliminating duplicate letters:")
print(set(result))
```

Exercise 9 MAP:
 Write a Python program to add two given lists and find the difference between them. Use the map() function


```python
def add_and_find_difference(list1,list2):
  sum1 = sum(map(int, list1))
  sum2 = sum(map(int, list2))

  difference = abs(sum1-sum2)

  return difference

list1 = ["1","2","3","4","5"]
list2 = ["3","4","5","8","5"]

result = add_and_find_difference(list1, list2)
print("Difference:", result)
```

Write a Python program to filter the height and weight of students, which are stored in a dictionary using lambda.

{'Cierra Vega': (6.2, 71), 'Alden Cantrell': (5.9, 65), 'Kierra Gentry': (6.0, 68), 'Pierre Cox': (5.8, 66)}

Exercise 14 Reduce:


Write a Python program to calculate the average value of the numbers in a given tuple of tuples using lambda.
Original Tuple:
((10, 10, 10), (30, 45, 56), (81, 80, 39), (1, 2, 3))
Average value of the numbers of the said tuple of tuples:
(30.5, 34.25, 27.0)


```python
import functools
original_tuple =  ((10,10,10),(30,45,56),(8,80,39),(1,2,3))

average_values = tuple(map(lambda x: functools.reduce(lambda a,b: a+b, x)/len(x), zip(*original_tuple)))

print("Original Tuple:")
print(original_tuple)

print("\nAverage value of the numbers of the said tuple of tuples:")
print(average_values)
```

Exercise 16:
Write a Python program to count the occurrences of items in a given list using lambda.
Original list:
[3, 4, 5, 8, 0, 3, 8, 5, 0, 3, 1, 5, 2, 3, 4, 2]
Count the occurrences of the items in the said list:
{3: 4, 4: 2, 5: 3, 8: 2, 0: 2, 1: 1, 2: 2}


```python
original_list = [3,4,5,8,0,3,1,5,2,3,4,2]

occurances = dict(map(lambda x:(x, original_list.count(x)), original_list))

print("Original list:")
print(original_list)

print("\nCount the occurances of items in the said list:")
print(occurances)
```
