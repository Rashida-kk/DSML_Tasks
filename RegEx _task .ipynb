1. Write a Python program to check that a string contains only a certain set of characters (in this case a-z, A-Z and 0-9).



```python
import re

txt = input("Enter the string: ")
x = re.compile(r'^[a-zA-Z0-9]+$')

if x.match(txt):
   print(f" '{txt}' contains only allowed characters")
else:
  print(f" '{txt}' contains characters outside the allow set")
```

2. Write a Python program that matches a string that has an a followed by zero or more b's.



```python
import re

def match_pattern(input_string):
  pattern = re.compile(r"^ab*$")

  if pattern.match(input_string):
    return True
  else:
    return False

test_string1 = "ab"
test_string2 = "abb"
test_string3 = "ac"

print(f"'{test_string1}' matches the pattern: {match_pattern(test_string1)}")
print(f"'{test_string2}' matches the pattern: {match_pattern(test_string2)}")
print(f"'{test_string3}' matches the pattern: {match_pattern(test_string3)}")
```

4. Write a Python program that matches a string that has an a followed by three 'b'.



```python
import re

text = input("Enter astring: ")
x = re.findall("a[b]{3}", text)

if x:
  print("Match found :", x)
else:
  print("Match not found :")
```

5. Write a Python program that matches a string that has an 'a' followed by anything ending in 'b'.

7. Write a Python program to search for literal strings within a string.\
Sample text : 'The quick brown fox jumps over the lazy dog.' \
Searched words : 'fox', 'dog', 'horse'

```python
import re

text =  'The quick brown fox jumps over the lazy dog.'
pattern_list = ['fox', 'dog', 'horse']

for pattern in pattern_list:
  if re.search(pattern, text):
    print(f"\'{pattern}\' is found in \'{text}\'")
  else:
    print(f"\'{pattern}\' is not found ")
```

8. Write a Python program to search for a literal string in a string and also find the location within the original string where the pattern occurs.

Sample text : 'The quick brown fox jumps over the lazy dog.' \
Searched words : 'fox'

```python
import re

pattern = 'fox'
text = 'The quick brown fox jumps over the lazy dog.'
x = re.search(pattern, text)

s = x.start()
e = x.end()
print('found "%s" in "%s" from %d to %d' %(x.re.pattern, x.string, s, e))
```

10. Write a Python program to find URLs in a string.



```python
import re

text = '<p>Contents :</p><a href="https://w3resource.com">Python Examples</a><a href="http://github.com">Even More Examples</a>'
x = re.findall(r'[ht]\w*://\w+.\w+',text)
print(x)
```

11. Write a Python program to remove the parenthesis area in a string.\

Sample data : ["example (.com)", "w3resource", "github (.com)", "stackoverflow (.com)"] \
Expected Output: \
example  \
w3resource \
github  \
stackoverflow. \



```python
import re

text = ["example(.com)", "w3resource", "github(.com)", "stackoverflow(.com)"]
for pattern in text:
  x = re.sub(r'\(\.\w+\)','',pattern)
  print(x)
```

12. Write a Python program to concatenate the consecutive numbers in a given string. \

Original string:  \
Enter at 1 20 Kearny Street. The security desk can direct you to floor 1 6. Please have your identification ready.  \

After concatenating the consecutive numbers in the said string:  \
Enter at 120 Kearny Street. The security desk can direct you to floor 16. Please have your identification ready.

```python
import re

original_string="Enter at 1 20 Kearny Street. The security desk can direct you to floor 1 6. Please have your identification ready."
x = re.compile(r'(?<=\d)( )(?=\d)')
res = x.sub('', original_string)
print(res)
```
