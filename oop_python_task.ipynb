1.  Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.

```python
import math

class Circle:
  def __init__(self,radius):
    self.radius = radius

  def calculate_area(self):
    return math.pi * self.radius **2

  def calculate_perimeter(self):
    return 2 * math.pi * self.radius

circle1 = Circle(10)

print("Area of circle =", circle1.calculate_area())
print("Perimeter of circle =", circle1.calculate_perimeter() )
```

```python
class Calculator:
  def __init__(self, num1, num2):
    self.num1 = num1
    self.num2 = num2

  def add(self):
    return self.num1 + self.num2

  def sub(self):
    return self.num1 - self.num2

  def mult(self):
    return self.num1 * self.num2

  def div(self):
    if self.num2 %2!= 0:
       return self.num1 / self.num2
    else:"cannot divided by 0"

calc = Calculator(10,5)

print("Sum of numbers =", calc.add())
print("Diference of numbers =", calc.sub())
print("Product of numbers =", calc.mult())
print("Quotient of numbers =", calc.div())
```
